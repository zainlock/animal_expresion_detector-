```python
import pandas as pd
```

```python
df=pd.read_csv("homeprices.csv")
df
```

## Create Dummies Variables with Pandas

dummies = pd.get_dummies(df.town)
dummies = dummies.astype(int)
dummies

```python
merge=pd.concat([df,dummies],axis='columns')
merge 
```

```python
final=merge.drop(['town','west windsor'],axis=1)
final
```

```python
x= final.drop(['price'],axis=1)
x
```

```python
y=final.price
y
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
# let's predict the price of the 2800 square fit in Robinsville
model.predict([[2800,0,1]])
```

```python
# let's predict the price of the 2800 square fit in west Windsor 
model.predict([[3400,0,0]])
```

## Create Dummies Variables Using sklearn

```python
df
```

```python
# need to Label encode from sklearn preprocessing 
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
delf=df
le.fit_transform(delf.town)
```

```python
delf.town=le.fit_transform(delf.town)
delf
```

```python
x=delf[['town','area']].values
x
```

```python
y=delf.price
y
```

```python
#create Dummy variables 
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
x=ct.fit_transform(x)
x
```

```python
x=x[:,1:]
x
```

```python
model.fit(x,y)
```

```python
# let's predict the price of the 2800 square fit in Robinsville
model.predict([[1,0,2800]])
```

```python
# let's predict the price of the 2800 square fit in west windstorm 
model.predict([[0,1,3400]])
```

```python
model.score(x,y)
```
